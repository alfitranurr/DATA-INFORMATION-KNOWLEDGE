```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
import seaborn as sns
from bs4 import BeautifulSoup
```

```python
continents_page = requests.get('https://simple.wikipedia.org/wiki/List_of_countries_by_continents').text
continents_page
```

```python
continents_countries_soup = BeautifulSoup(continents_page, 'html.parser')
continents = continents_countries_soup.find_all('div' > 'h2', {"class":'mw-heading'})
continents
```

```python
unwanted_words = ['Sovereign states', 'Dependent territories', 'Notes', 'References', 'Other website']
target_continents = [continent.text for continent in continents if continent.text not in unwanted_words]
target_continents
```

```python
td_html = continents_countries_soup.find_all('tr')
all_countries = [countries.find_all('td', {"class":None, "id": None}) for countries in td_html]
all_countries
```

```python
countries_in_continents = []
for items in all_countries:
    countries = []
    if items:
        # Menyaring hanya elemen yang mengandung tag <a> dan mengambil teksnya
        countries = [country.find('a').text for country in items if country.find('a')]
    
    # Ambil hanya elemen pertama dari setiap daftar jika ada data
    if countries:
        countries_in_continents.append(countries[1])

countries_in_continents
```

## World Happiness Report - 2020

```python
countries_score_page = requests.get('https://en.wikipedia.org/wiki/World_Happiness_Report#2020_report')
countries_score_soup = BeautifulSoup(countries_score_page.content, 'lxml')
```

```python
countries_score_table = countries_score_soup.find_all('table', {'class': 'wikitable'})
countries_score_table
```

```python
from io import StringIO

# Wrap the HTML content with StringIO
html_content = str(countries_score_table)
countries_score_df = pd.read_html(StringIO(html_content))
countries_score_df
```

```python
countries_score_df = countries_score_df[0]
countries_score_df = countries_score_df.rename(columns={"Country or region": "Country"} )
countries_score_df
```
